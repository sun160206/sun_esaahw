```
from pandas import DataFrame
# 데이터 만들기
ex={'경도':[126.94534071256834,126.94481815956661,126.94367131391576,126.94478693400981,126.94407646653858], 
    '위도':[37.55860610691874,37.55947113742418,37.558635617737025,37.558488169490545,37.559153145013894], 
    '상호명':['버거앤치즈','리또리또','배러댄와플','아웃닭','벨라프라하']}

ex=DataFrame(ex)
ex
```

```
import folium

#지도 중심 지정하기 위해 위도와 경도 평균 구하기
lat=ex['위도'].mean()
long=ex['경도'].mean()

#지도 띄우기
m=folium.Map([lat,long],zoom_start=20)
```

```
for i in ex.index:
  sub_lat = ex.loc[i,'위도']
  sub_long =ex.loc[i,'경도']

  title=ex.loc[i,'상호명']

  # 지도에 데이터 찍어서 보여주기
  folium.Marker([sub_lat,sub_long],tooltip=title).add_to(m)


#한글 안나오는 오류로 html로 trouble shooting
m.save('example.html')
m
```
